```python
# Definição da função analisar_credito
def analisar_credito(score, renda):
    if score >= 700 and renda >= 5000:
        return "Crédito aprovado com juros de 2% ao mês"
    elif score >= 600 and renda >= 3000:
        return "Crédito aprovado com juros de 4% ao mês"
    else:
        return "Crédito negado"

# Função para testar a lógica de crédito
def testar_credito():

    assert analisar_credito(750, 6000) == "Crédito aprovado com juros de 2% ao mês"
    assert analisar_credito(650, 3500) == "Crédito aprovado com juros de 4% ao mês"
    assert analisar_credito(500, 2000) == "Crédito negado"
    print("✅ Todos os testes passaram!")

# Executando os testes
testar_credito()
```
