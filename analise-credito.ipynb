```python
# Função para analisar crédito com base no score e renda
def analisar_credito(score, renda):
    """
    Parâmetros:
    - score (int): Pontuação de crédito do cliente (0 a 1000).
    - renda (float): Renda mensal do cliente.

    Retorna:
    - str: Resultado da análise de crédito.
    """
    if score >= 700 and renda >= 5000:
        return "Crédito aprovado com juros de 2% ao mês"
    elif score >= 600 and renda >= 3000:
        return "Crédito aprovado com juros de 4% ao mês"
    else:
        return "Crédito negado"

# Exemplo de uso
clientes = [
    {"nome": "Ana", "score": 750, "renda": 6000},
    {"nome": "Bruno", "score": 650, "renda": 3500},
    {"nome": "Carlos", "score": 500, "renda": 2000}
]

for cliente in clientes:
    resultado = analisar_credito(cliente["score"], cliente["renda"])
    print(f"Cliente: {cliente['nome']} | Resultado: {resultado}")
```
